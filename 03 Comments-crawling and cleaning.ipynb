Crawling comments under videos on Bibili

```python
#import functions from external packages
import requests
import time
from bs4 import BeautifulSoup
import json

#define a function to crawl the page
def get_html(url):
#make a get request for the page
    headers = {
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36',
    }
    r = requests.get(url, timeout=30,headers=headers)
    r.raise_for_status()
    r.endcodding = 'utf-8'
    return r.text

#define a function to get contents on pages
def get_content(url):
#define a list to save data
    comments = []
    html = get_html(url)
    try:
        s=json.loads(html)
    except:
        print("jsonload error")  
    num=len(s['data']['replies']) 
    # print(num)
    i=0
    while i<num:
        comment=s['data']['replies'][i]
#create a dictionary
        InfoDict={}       
        InfoDict['Uname']=comment['member']['uname'] 
        InfoDict['Like']=comment['like'] 
        InfoDict['Content']=comment['content']['message'] 
        InfoDict['Time']=time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(comment['ctime'])) #change into ctime version
        
        comments.append(InfoDict)
        i=i+1
# print(comments)
    return comments

#put results into csv
def Out2File(dict):
    with open('RAWComments-video2.csv', 'a+',encoding='utf-8') as f:
        i=0
        for comment in dict:
            i=i+1
            try:
                f.write('姓名：{}\t\n点赞数：{}\t\n评论内容：{}\t\n评论时间：{}\t\n '.format(
                    comment['Uname'], comment['Like'], comment['Content'], comment['Time']))
                f.write("-----------------\n")
            except:
                print("out2File error")
        print('当前页面保存完成')

if __name__ == '__main__':
    e=0
    page=1
    while e == 0 :
        url = "https://api.bilibili.com/x/v2/reply?pn="+ str(page)+"&type=1&oid=89729997&sort=2" 
        try:
            print()
            content=get_content(url)
            print("page:",page)
            Out2File(content)
            page=page+1
            if page%10 == 0:
                time.sleep(5)
        except:
            e=1
```

```python
#import functions from external packages
#read data from files and turn to dataframe
import pandas as pd
import numpy as np
with open('RAWComments-video2.csv',encoding='utf-8') as f:
    lst=f.read().splitlines()
res=[]
row={}
NAME_,LIKES_,DATE_,CONTENT_='姓名：','点赞数：','评论时间：','评论内容：'
for s in lst:
    if s==' -----------------':
        res.append(row)
        row['评论内容']='\n'.join(row['评论内容'])
        row={}
    elif s.startswith(NAME_)  :
        row['姓名']=s.strip().replace(NAME_,'')  
    elif s.lstrip().startswith(LIKES_):
        row['点赞数']=s.strip().replace(LIKES_,'')
    elif s.lstrip().startswith(DATE_):
        row['评论时间']=s.strip().replace(DATE_,'')
    elif s.lstrip().startswith(CONTENT_):
        row['评论内容']=[s.lstrip().replace(CONTENT_,'')]
    else:
        row['评论内容'].append(s)
df = pd.DataFrame(res)
#select data from a specific period
df['评论时间'] = pd. to_datetime(df['评论时间']).dt.normalize()
df = df.set_index('评论时间')
#here take first video as an example
#time period of other four videos:2020-02-17 to 2020-02-21/2020-02-21 to 2020-02-24/2020-02-24 to 2020-02-26/2020-02-26 to 2020-02-28
print(df['2020-02-17':'2020-02-21'])
df = df['2020-02-17':'2020-02-21']
df.to_csv('Comments-video2.csv',encoding='utf-8',index=True)
df = pd.read_csv('Comments-video2.csv',encoding='utf-8')
df
```

Data cleaning

```python
df = df.copy()
df = df.drop_duplicates() #delete repeating lines
df.columns = ["time","id","like","comment"] #rename
def yasuo(st):
    for i in range(1,int(len(st)/2)+1):
        for j in range(len(st)):
            if st[j:j+4] == st[j+i:j+4*i]:
                k = j + i
                while st[k:k+i] == st[k+i:k+2*i] and k<len(st):   
                    k = k + i
                st = st[:j] + st[k:]    
    return st
yasuo(st="钉钉钉钉")
df["comment"] = df["comment"].apply(yasuo)
import re
def clean(text):
    text = re.sub(r"\[\S+\]", "", text)      # delete special characters
    return text.strip()
df["comment"] = df["comment"].apply(clean)
df = df.dropna()
df
```
